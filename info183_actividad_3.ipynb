#### Enunciado

En esta tarea se pide entrenar y evaluar un predictor para la serie de tiempo Mackey-Glass. Esta serie de tiempo se obtiene de la solución de la siguiente ecuación diferencial

$$
\frac{dy}{dt} = 0.2 \frac{ y(t-\tau)}{1 + y(t-\tau)^{10}} - 0.1 y(t),
$$

donde el parámetro $\tau$ controla el comportamiento dinámico de la serie de tiempo 

***

Use el código que se muestra a continuación para generar la serie de tiempo Mackey Glass

- Considere una razón señal a ruido (SNR) de 2.
- Considere $\tau=17$ (comportamiento debilmente caótico)

Se generaran 1000 muestras de la serie de tiempo. Use los primeros 500 puntos para entrenar y los siguientes 500 puntos para hacer predicción

```python
# número de observaciones (no modificar)
N = 1000
# Razón señal a ruido (2., 0.5)
SNR = 2.
# constantes de la ecuación diferencial (no modificar)
a, b = 0.1, 0.2
# comportamiento dinámico de Mackey-Glass (17, 30)
tau = 17.
# paso de integración (no modificar)
dt = 0.05
# condición incial (no modificar)
y0 = 0.9
# largo temporal (no modificar)
tt = 5.
t = np.linspace(0, tt, num=N)

N_full, tau_full = int(N*tt/dt), int(tau/dt) #Adapta los índices a la segmentación
#print(tau_full, N_full) -> 340, 100.000
ymg = y0*np.ones(shape=(N_full, ))
# Runge-Kutta integration
for n in range(tau_full, N_full-1):
    byd = b*ymg[n-tau_full]/(1.0 + ymg[n-tau_full]**10.0)
    yk1 = dt*(-a*ymg[n] + byd) # Derivada del enunciado
    yk2 = dt*(-a*(ymg[n]+yk1/2) + byd)
    yk3 = dt*(-a*(ymg[n]+yk2/2) + byd)
    yk4 = dt*(-a*(ymg[n]+yk3) + byd)
    ymg[n+1] = ymg[n] + yk1/6 + yk2/3 +yk3/3 +yk4/6; # Predicción Ponderada (n+1)
ymg = ymg[::int(tt/dt)]
#ymg = ymg - np.mean(ymg)

# Contaminación con ruido blanco aditivo:
s_noise = np.sqrt(np.var(ymg)/SNR) 
np.random.seed(0)
y_obs = ymg + s_noise*np.random.randn(len(ymg))

# Gráfico
fig, ax = plt.subplots(1, figsize=(9, 3), tight_layout=True)
ax.plot(t[:500], y_obs[:500])
ax.set_title('Serie de tiempo Mackey-Glass (entrenamiento)');
```

Respuestas:

Semejanzas:
- La forma de optimizar su funcionalidad, consta de optimizar la función de costo J, que dependen del error entre la señal predecida, y la real.
- El funcionamiento se basa en ajustar los valores (o 'pesos') de una matriz 'w'.
- Dependen de una señal esperada 'd'.
    
Diferencias:
- El filtro de Weiner tiene una estructura LTI, mientras que el algoritmo LSM resulta en un filtro FIR.
- Para la función costo J del Weiner, se ocupa como criterio el MSE (error cuadrático medio), mientras que para el LMS se ocupa el SGD (gradiente descendiente estocástico).
- En sí, el LMS es una extensión de Weiner, que sí permite casos no-estacionarios. Este último asume que estacionariedad ( J(h) = E[$e_n^2$] ).
- El orden de complejidad para Weiner es de O(L²), relacionado con un ajuste determinista, mientras que LMS tiene complejidad O(L), con un ajuste estadístico.
- Weiner es óptimo, pero no adaptativo. LMS es adaptativo (de hecho, puede actualizarse en tiempo real), pero no óptimo, aunque sí tiende a la solución óptima de Weiner.

2)<br>
Primero:<br>
$J_n^s(w) = e_n^2$<br>
= $(d_n - y_n)^2$<br>
= $(d_n - w^T u_n)^2$<br>
$$ = (d_n - \sum_{k=0}^L w_{n,k} u_{n-k})^2$$<br>
 
así, su derivada es:<br>
$$\frac{dJ_n^s(w)}{dw_{n,k}} = -2e_n u_{n-k}$$<br>

finalmente, junto a la regla SGD (utilizando $\mu$):<br>
$w_{n+1,k} = w_{n,k} + \mu * 2 e_n u_{n-k}$<br>

y de forma matricial:<br>
$w_{n+1} = w_n + \mu * 2 e_n u_n$<br>
$w_{n+1} = w_n + \mu * 2(d_n - w_n^T u_n)u_n$<br>

Para corregir dominancia en vector $u_n$, lo último se extiende con el algoritmo NLMS:<br>
$w_{n+1} = w_n + \mu * 2 e_n \frac{u_n}{||u_n||^2 + \delta}$<br>

3) Dado a que la clase incluye una corrección para la norma, será ocupada la última ecuación de la sección anterior. Así, la línea deberá ser:

self.w = self.w + 2 * self.mu * (d_pred - np.transpose(self.w) * u) * u / norm 


```python
class LMS_filter(object):
    
    def __init__(self, L=1, mu=0.5, normalized=True):
        self.L = L
        self.mu = mu
        self.w = np.zeros(shape=(L, ))
        self.normalized = normalized
    
    def __len__(self):
        return self.L
    
    def predict(self, u):
        return np.dot(self.w, u)
    
    def update(self, u, d):
        d_pred = self.predict(u)
        norm = 1.
        if self.normalized:
            norm = np.sum(u**2) + 1e-6
        self.w = self.w + 2 * self.mu * (d_pred - np.transpose(self.w) * u) * u / norm 
```

1. Describa en detalle el algoritmo RLS indicando sus semejanzas y diferencias con el algoritmo LMS
1. Partiendo del error histórico $J_N(\textbf{w}) = \sum_{i=1}^N \beta^{N-i} e_i^2$ derive la regla recursiva de actualización de pesos 
1. La siguiente clase de *Python* predice y entrena un filtro RLS. Complete las líneas que dice 

` self.w = ? ` y `self.Phi_inv = ?`

con el valor correcto de actualización de peso del filtro RLS

Respuestas:<br>
1) <br>

2) <br>
    $J_N(\textbf{w}) = \sum_{i=1}^N \beta^{N-i} e_i^2$<br>
    $$= \sum_{i=L+1}^n \beta^{n-i} (d_i - \sum_{k=0}^{L} w_{i, k} u_{i-k} )^2$$<br>
    Agregando  $J^w_n = \delta  \| \textbf{w}_{n} \|^2$:<br>
    $$\textbf{w}_n = (U_n^T \pmb{\beta} U_n + \delta I)^{-1}  U_n^T \pmb{\beta} \textbf{d}_n \tag{1}$$<br>
    Como RLS se actualiza de manera recursiva, se tiene que actualizar para $\Phi_n^{-1}$ es mas eficiente que $\Phi_n$. Así:       <br>
    $\Phi_{n} = \beta \Phi_{n-1} + \textbf{u}_n \textbf{u}_n^T\tag{2}$<br>
    $\theta_{n} = \beta \theta_{n-1} + \textbf{u}_n d_n \tag{3} $ <br>
   $$\Phi_{n}^{-1}= \beta^{-1} \Phi_{n-1}^{-1} - \beta^{-1} \textbf{k}_n \textbf{u}_n^T \Phi_{n-1}^{-1}\tag{4}$$<br>
   De esta manera, el cálculo de los pesos queda como: <br>
   $$
    \begin{align}
    \textbf{w}_n &= \Phi_n^{-1} \theta_n \tag{5} \\
    .\\
    .\\
    .\\
&=  \textbf{w}_{n-1} + \textbf{k}_n e_n \tag{6}
\end{align}
  $$
  donde $\textbf{k}_n$(*ganancia*)$ = \left(\beta^{-1} \Phi_{n-1}^{-1} - \beta^{-1} \textbf{k}_n \textbf{u}_n^T \Phi_{n-1}^{-1} \right)  \textbf{u}_n = \Phi_n^{-1} u_n$<br>
3) Tomando en cuenta las ecuaciones (4) y (6), self.w y self.Phi_inv serían, respectivamente:<br>

```python
class RLS_filter(object):
    
    def __init__(self, L=1, beta=0.9, delta=10.):
        self.L = L
        self.beta = beta
        self.w = np.zeros(shape=(L, ))
        self.Phi_inv = delta*np.eye(L)
    
    def __len__(self):
        return self.L
    
    def predict(self, u):
        return np.dot(self.w, u)
    
    def update(self, u, d):          
        invbeta = 1.0/self.beta
        d_pred = self.predict(u)
        e = d - d_pred
        r = 1. + invbeta*np.dot(np.dot(u, self.Phi_inv), u.T)
        k = invbeta*np.dot(self.Phi_inv, u)/r       
        self.Phi_inv = (invbeta * self.Phi_inv) - (invbeta * (self.Phi_inv*u) * np.transpose(u) * self.Phi_inv)
        self.w = self.w + k*e
        
```
